```python
import pandas as pd 
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# read the data

```python
movies_dataset = pd.read_csv('movie_dataset.csv')
```

```python
movies_dataset.head()
```

```python
movies_dataset.columns
```

# feature selection 

```python
features  = ['keywords', 'cast', 'genres', 'director']
```

```python
movies_dataset[features]
```

```python
def combine_feature(row):
#     try opertation
    try:
        return row['keywords']+' '+ row['cast']+' '+row['director']+' '+row['genres']
#     
    except:
        print('Error',row)
```

```python
combine_feature(features)
```

```python
movies_dataset['combine_features'] = movies_dataset.apply(combine_feature, axis=1)
```

```python
for feature in features:
    movies_dataset[feature] = movies_dataset[feature].fillna('')
```

```python
movies_dataset.isna().sum()
```

```python
movies_dataset['combine_features'].head()
```

```python
movies_dataset.head()
```

# create count matrix

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(movies_dataset['combine_features'])
```

```python
print(count_matrix)
```

# cosine similarity

```python
cosine_similarity = cosine_similarity(count_matrix)
```

```python
cosine_similarity.shape
```

```python
movie_user_likes = 'Avatar'
```

```python
###### helper functions. Use them when needed #######
def get_title_from_index(index):
    try:
        return movies_dataset[movies_dataset.index == index]["title"].values[0]
    except:
        return "error", index
    
    
    
def get_index_from_title(title):
	return movies_dataset[movies_dataset.title == title]["index"].values[0]
```

```python
movie_index = get_index_from_title(movie_user_likes)
```

```python
movie_index
```

```python
similar_movies = list(enumerate(cosine_similarity[movie_index]))
```

```python
similar_movies
```

```python
sorted_similar_movies = sorted(similar_movies, key= lambda x:x[1], reverse=True)
```

```python
i=0
for movie in sorted_similar_movies:
    print (get_title_from_index(movie[0]))
    i=i+1
    if i>50:
        break
```
