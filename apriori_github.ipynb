```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('data.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.columns = map(str.lower, df.columns)
```

```python
df.columns
```

```python
df.head()
```

```python
df['description'] = df['description'].str.strip()
df.dropna(axis=0, subset = ['invoiceno'], inplace=True)
df['invoiceno']  = df['invoiceno'].astype('str')
df = df[-df['invoiceno'].str.contains('C')]
df
```

```python
# make basket with frequency for any item in each purchase
basket = (df[df['country'] == 'France']
         .groupby(['invoiceno', 'description'])['quantity']
         .sum().unstack().reset_index().fillna(0)
         .set_index('invoiceno'))
```

```python
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
baskets_sets  = basket.applymap(encode_units)         ##encode the basket

baskets_sets.drop('POSTAGE',axis=1, inplace = True)
baskets_sets
```

```python
frequent_itemsets = apriori(baskets_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift',min_threshold=1)
rules.head()
```
